```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/dsrscientist/dataset3/main/glass.csv')
```

```python
df.head()
```

```python
df.rename(columns={'1':'ID Number','1.52101':'RI','13.64':'Na','4.49':'Mg','1.10':'Al','71.78':'Si','0.06':'K','8.75':'Ca','0.00':'Ba','0.00.1':'Fe','1.1':'Type of Glass'}, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
for feat in df:
    skew = df[feat].skew()
    sb.distplot(df[feat], kde= False, label='Skew = %.3f' %(skew), bins=30)
    plt.legend(loc='best')
    plt.show()
```

Fe,Ba,Ca,K - skew rate high
K,Ba - more outliers

```python
plt.figure(figsize=(8,8))
sb.pairplot(df,palette='coolwarm')
plt.show()
```

```python
corr=df.corr()
plt.figure(figsize=(10,10))
sb.heatmap(corr,cbar=True, square=True, annot=True, fmt='.2f',annot_kws={'size': 10},xaxis= df, yaxis= df, alpha= 0.7, cmap='coolwarm')
plt.show()
```
